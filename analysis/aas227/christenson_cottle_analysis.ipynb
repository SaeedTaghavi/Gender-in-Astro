We categorize oral sessions at AAS227 into the following categories, based on the six astrophysics categories used by Arxiv:

1. Galaxies: Galaxies, including the Milky Way, and their contents. Galactic structure and evolution, objects within galaxies, ISM ,clouds, and dust.

2. Cosmology: Non-galactic cosmology including CMB, extragalactic distance scale, intergalactic medium, large-scale structure of the universe, particle astrophysics.

3. High Energy: Cosmic rays, gamma and x-ray astronomy, supernovae, black holes, AGN, neutron stars, pulsars.

4. Earth/Planetary: Astrobiology, exoplanet detection and properties, structure and formation of the solar system.

5. Solar/Stellar: Star formation, binary systems, white and brown dwarfs, stellar evolution and structures.

6. Methods: Data analysis, statistics, catalogs, surveys, instrumentation, and software.

7. Education, Outreach, and Professional Development: History of astronomy, astronomy education research, public outreach, teaching practices, and methodology lectures. Includes all plenary sessions and town halls.


We assign the numbers 1-7 to each category as a category flag (as listed above) to be added to the data set for each session. We sort the oral sessions into these categories and create lists of the sessions in each category below:

```python
#read in the catalog and define an array for each column
gender_data = ascii.read('genderinastro_data_revised.csv')

session = gender_data['session']
chair = gender_data['chair']
speaker = gender_data['speaker']
questions = gender_data['questions']
```

```python
#define an array for categories
category = np.zeros(np.size(session))

#need to round session numbers down to eliminate decimal points:
session_rounded = np.round(session,0)
session_rounded = session_rounded.astype(int)

for i in range(0,np.size(session)):
    if session_rounded[i] in category1:
        category[i] = 1
    if session_rounded[i] in category2:
        category[i] = 2
    if session_rounded[i] in category3:
        category[i] = 3
    if session_rounded[i] in category4:
        category[i] = 4
    if session_rounded[i] in category5:
        category[i] = 5
    if session_rounded[i] in category6:
        category[i] = 6
    if session_rounded[i] in category7:
        category[i] = 7
category = category.astype(int)

#print category
```

```python
#append category array onto gender_data table

col=Column(data=category,name='Category')
gender_data.add_column(col)

print gender_data
```

```python
#write table as a file

ascii.write(gender_data,'aas_gender_data.csv')
```
